```python
# ============================================
# 📦 IMPORTS
# ============================================
import numpy as np
import pandas as pd
from scipy import stats
from sklearn.datasets import load_iris
import seaborn as sns

# ============================================
# 📥 LOAD DATASETS
# ============================================

# Iris Dataset
iris = load_iris(as_frame=True)
iris_df = iris.frame

# Titanic Dataset
titanic_df = sns.load_dataset("titanic")

print("Datasets Loaded ✅")
print(iris_df.head())
print(titanic_df.head())


# ============================================
# 🟢 1. Z-TEST (Manual)
# ============================================

print("\n===== Z-TEST =====")

data = iris_df['sepal length (cm)']

sample_mean = np.mean(data)
population_mean = 5.5   # assumed
population_std = np.std(data)
n = len(data)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

print("Z-score:", z_score)


# ============================================
# 🟡 2. T-TEST
# ============================================

print("\n===== T-TEST =====")

# 1-Sample T-Test
t_stat, p_val = stats.ttest_1samp(data, popmean=5.5)
print("1-Sample T-test:", t_stat, p_val)

# 2-Sample T-Test (setosa vs versicolor)
setosa = iris_df[iris_df['target'] == 0]['sepal length (cm)']
versicolor = iris_df[iris_df['target'] == 1]['sepal length (cm)']

t_stat2, p_val2 = stats.ttest_ind(setosa, versicolor)

print("2-Sample T-test:", t_stat2, p_val2)


# ============================================
# 🔵 3. CHI-SQUARE TEST
# ============================================

print("\n===== CHI-SQUARE TEST =====")

# Create contingency table
contingency_table = pd.crosstab(titanic_df['sex'], titanic_df['survived'])

chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("Chi2:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)


# ============================================
# 🔴 4. ANOVA
# ============================================

print("\n===== ANOVA =====")

# Compare sepal length across species
group1 = iris_df[iris_df['target'] == 0]['sepal length (cm)']
group2 = iris_df[iris_df['target'] == 1]['sepal length (cm)']
group3 = iris_df[iris_df['target'] == 2]['sepal length (cm)']

f_stat, p_val_anova = stats.f_oneway(group1, group2, group3)

print("F-statistic:", f_stat)
print("p-value:", p_val_anova)


# ============================================
# 🧠 INTERPRETATION HELPER
# ============================================

alpha = 0.05

def interpret(p):
    return "Reject H0 ❌" if p < alpha else "Fail to Reject H0 ✅"

print("\n===== INTERPRETATIONS =====")

print("T-test:", interpret(p_val))
print("2-sample T-test:", interpret(p_val2))
print("Chi-square:", interpret(p))
print("ANOVA:", interpret(p_val_anova))
```
