Dumping Car Data into table "Car"

```python
import csv
with open("car.data", "r") as f:
    cardata = csv.reader(f, delimiter="\n")
    for i in cardata:
        cursor.execute("INSERT into cardataset.car VALUES(%s, %s, %s, %s, %s, %s, %s)",i[0].split(','))
```
