```python
from tkinter import *
from tkinter import messagebox, simpledialog
import math

class Calculator:
    def __init__(self,root):
        self.root = root
        self.root.title("Calculator")
        self.root.geometry("250x350+400+150")
        self.root.resizable(False,False)
        self.root.config(bg = "#864c7f")
        self.frame = Frame(self.root,relief="sunken",bd=10)
        self.frame.grid(row=0,column=0,columnspan=5,sticky=N+S+E+W,padx=2,pady=2)
        self.history = Label(self.frame,font = "Ariel 10 bold",relief = "flat",bg= "lightyellow",fg="grey",height = 2,anchor = "e")
        self.screen = Entry(self.frame, text = "0",font = "Calibre 15 bold",relief = "flat",bg= "lightyellow",width = 10,bd=5)
        self.b1 = Button(self.root, text = "1", font = "TimesNewRoman 15 bold",width=3,command = lambda: self.screen.insert(END,"1"),relief ="groove" )
        self.b2 = Button(self.root, text = "2", font = "TimesNewRoman 15 bold",width=3,command = lambda: self.screen.insert(END,"2"),relief ="groove" )
        self.b3 = Button(self.root, text = "3", font = "TimesNewRoman 15 bold",width=3,command = lambda: self.screen.insert(END,"3"),relief ="groove" )
        self.b4 = Button(self.root, text = "4", font = "TimesNewRoman 15 bold",width=3,command = lambda: self.screen.insert(END,"4"),relief ="groove" )
        self.b5 = Button(self.root, text = "5", font = "TimesNewRoman 15 bold",width=3,command = lambda: self.screen.insert(END,"5"),relief ="groove" )
        self.b6 = Button(self.root, text = "6", font = "TimesNewRoman 15 bold",width=3,command = lambda: self.screen.insert(END,"6"),relief ="groove" )
        self.b7 = Button(self.root, text = "7", font = "TimesNewRoman 15 bold",width=3,command = lambda: self.screen.insert(END,"7"),relief ="groove" )
        self.b8 = Button(self.root, text = "8", font = "TimesNewRoman 15 bold",width=3,command = lambda: self.screen.insert(END,"8"),relief ="groove" )
        self.b9 = Button(self.root, text = "9", font = "TimesNewRoman 15 bold",width=3,command = lambda: self.screen.insert(END,"9"),relief ="groove" )
        self.b0 = Button(self.root, text = "0", font = "TimesNewRoman 15 bold",width=3,command = lambda: self.screen.insert(END,"0"),relief ="groove" )
        self.bdot = Button(self.root, text = ".", font = "TimesNewRoman 15 bold",width=3,command = lambda: self.screen.insert(END,"."),bg="lightblue",relief ="groove" )
        self.bequal = Button(self.root, text = "=", font = "TimesNewRoman 15 bold",width=3,command = self.result,bg="lightblue",relief ="groove")
        self.bplus = Button(self.root, text = "+", font = "TimesNewRoman 15 bold",width=3,command = self.add,bg="#55fdf9",relief ="groove")
        self.bminus = Button(self.root, text = "-", font = "TimesNewRoman 15 bold",width=3,command = self.sub,bg="#55fdf9",relief ="groove")
        self.bdivide = Button(self.root, text = "/", font = "TimesNewRoman 15 bold",width=3,command = self.divide,bg="#55fdf9",relief ="groove")
        self.bmulti = Button(self.root, text = "x", font = "TimesNewRoman 15 bold",width=3,command = self.multi,bg="#55fdf9",relief ="groove")
        self.bsquare = Button(self.root, text = "x²", font = "TimesNewRoman 15 bold",width=3,command = self.square,bg="#55fdf9",relief ="groove")
        self.bsqrt = Button(self.root, text = "√", font = "TimesNewRoman 15 bold",width=3,command = self.sqrt,bg="#55fdf9",relief ="groove")
        self.bpower = Button(self.root, text = "xⁿ", font = "TimesNewRoman 15 bold",width=3,command = self.power,bg="#55fdf9",relief ="groove")
        self.braised = Button(self.root, text = "ⁿ√", font = "TimesNewRoman 15 bold",width=3,command = self.raised_to,bg="#55fdf9",relief ="groove")
        self.c = Button(self.root, text = "C", font = "TimesNewRoman 12 bold",width=3,command = self.backspace,bg="lightblue",relief ="groove")
        self.ce = Button(self.root, text = "CE", font = "TimesNewRoman 12 bold",width=3,command = self.clearall,bg="red",relief ="groove")
       
        self.history.pack(fill=X)
        self.screen.pack(fill=X)
        self.c.grid(row = 2, column =3,sticky = W+E,padx = 5,pady = 10)
        self.ce.grid(row=2,column=4,sticky = W+E,padx = 2,pady = 2)
        self.bsquare.grid(row=3,column=0,sticky = W,padx = 2,pady = 2)
        self.b1.grid(row=3,column=1,sticky = W,padx = 2,pady = 2)
        self.b2.grid(row=3,column=2,sticky = W,padx = 2,pady = 2)
        self.b3.grid(row=3,column=3,sticky = W,padx = 2,pady = 2)
        self.bminus.grid(row=3,column=4,sticky = W,padx = 2,pady = 2)
        self.bsqrt.grid(row=4,column=0,sticky = W,padx = 2,pady = 2)
        self.b4.grid(row=4,column=1,sticky = W,padx = 2,pady = 2)
        self.b5.grid(row=4,column=2,sticky = W,padx = 2,pady = 2)
        self.b6.grid(row=4,column=3,sticky = W,padx = 2,pady = 2)
        self.bmulti.grid(row=4,column=4,sticky = W,padx = 2,pady = 2)
        self.bpower.grid(row=5,column=0,sticky = W,padx = 2,pady = 2)
        self.b7.grid(row=5,column=1,sticky = W,padx = 2,pady = 2)
        self.b8.grid(row=5,column=2,sticky = W,padx = 2,pady = 2)
        self.b9.grid(row=5,column=3,sticky = W,padx = 2,pady = 2)
        self.bdivide.grid(row=5,column=4,sticky = W,padx = 2,pady = 2)
        self.braised.grid(row=6,column=0,sticky = W,padx = 2,pady = 2)
        self.bdot.grid(row=6,column=1,sticky = W,padx = 2,pady = 2)
        self.b0.grid(row=6,column=2,sticky = W,padx = 2,pady = 2)
        self.bequal.grid(row=6,column=3,sticky = W,padx = 2,pady = 2)
        self.bplus.grid(row=6,column=4,sticky = W,padx = 2,pady = 2)

    def add(self):
        self.screen.insert(END,"+")

    def sub(self):
        self.screen.insert(END,"-")

    def multi(self):
        self.screen.insert(END,"*")

    def divide(self):
        self.screen.insert(END,"/")

    def result(self):
        try:
            rslt = eval(self.screen.get())
            self.history["text"] = self.screen.get()
            self.screen.delete(0,END)
            self.screen.insert(0,rslt)
        except ZeroDivisionError:
            messagebox.showerror("Arithmetic Error","Division by zero")
        except NameError:
            messagebox.showerror("Value Error","Only Digits are allowed")
        except SyntaxError:
            messagebox.showerror("Syntax Error","No input!!")

    def clearall(self):
        self.screen.delete(0,END)
        self.history["text"] = ""

    def backspace(self):
        self.screen.delete(first=END)
        
    def square(self):
        rslt = float(self.screen.get())**2
        self.history["text"] = self.screen.get()+"^2"
        self.screen.delete(0,END)
        self.screen.insert(0,str(rslt))
        
        
    def sqrt(self):
        rslt = math.sqrt(float(self.screen.get()))
        self.history["text"] = "√"+self.screen.get()
        self.screen.delete(0,END)
        self.screen.insert(0,str(rslt))
        
    def power(self):
        a,b=self.screen.get().split("^")
        rslt = math.pow(float(a.strip()),float(b.strip()))
        self.history["text"] = self.screen.get()+"^"+b
        self.screen.delete(0,END)
        self.screen.insert(0,str(rslt))
        
    def raised_to(self):
        a,b=self.screen.get().split("√")
        rslt = math.pow(float(b.strip()),1/float(a.strip()))
        self.history["text"] = b+"√"+self.screen.get()
        self.screen.delete(0,END)
        self.screen.insert(0,str(rslt))
```

```python
from tkinter import *
from tkinter import messagebox, simpledialog
from tkinter import ttk as ttk
class Weight_Conversion:
    def __init__(self,root):
        self.root=root
        self.root.title("Weight Conversion")
        self.root.geometry("300x220+400+150")
        self.root.resizable(False,False)
        self.root.config(bg="#f6b26b")
        style = ttk.Style(root)
        style.configure("my.TMenubutton", font= ("Calibre", 10, "bold","italic"),relief = "raised",width = 15)
        options = ["Tons(t)","Tons(t)","Pounds(lb)","Ounces(oz)","Kilogrammes(kg)","Grams(g)","Quintal"]
        self.clicked = StringVar()
        self.dp1 = ttk.OptionMenu(root,self.clicked,*options,style = "my.TMenubutton")
        self.clicked2 = StringVar()
        self.dp2 = ttk.OptionMenu(root,self.clicked2,*options,style = "my.TMenubutton")
        self.lbl = Label(root,text = "Weight Unit Converter",font= "Times 20 bold", bg = "lightblue",fg = "red",relief="raised",bd=5)
        self.screen1 = Entry(root,text="1",font = "Calibre 20 bold",relief = "sunken",bg= "lightyellow",width = 15,bd=5)
        self.screen2 = Label(root,font = "Calibre 20 bold",relief = "solid",bg= "lightyellow",width = 15,anchor="w")
        self.lbl1 = Label(root,text="From",relief = "flat",anchor="w",font=("Calibre",10,"bold"),bg="#f6b26b")
        self.lbl2 = Label(root,text="To",relief = "flat",anchor="w",font=("Calibre",10,"bold"),bg="#f6b26b")
        ttk.Style(root).configure("my.TButton",font = ("Ariel",7,"bold"))
        self.bconv = ttk.Button(self.root,text = "Convert", style="my.TButton",command = self.convert)
        self.clear = ttk.Button(self.root,text = "Clear", style="my.TButton",command = self.clearall)
        self.lbl.grid(row = 0, column=0,columnspan = 3,sticky= W+E,padx=7)
        self.lbl1.grid(row=1,column = 0,sticky = W)
        self.dp1.grid(row = 1 ,column = 2,sticky = W)
        self.screen1.grid(row = 2, column=0,columnspan = 3,sticky= W+E,padx=5,pady=5)
        self.lbl2.grid(row=3,column = 0,sticky = W)
        self.dp2.grid(row = 3 ,column = 2,sticky = W)
        self.screen2.grid(row = 4, column=0,columnspan = 3,sticky= W+E,padx=5,pady=5)
        self.bconv.grid(row=5,column = 0,sticky = W,padx = 1,pady=3)
        self.clear.grid(row=5,column = 1,sticky = W,padx = 1,pady=3)
        
    def clearall(self):
        self.screen1.delete(0,END)
        self.screen2["text"]=""
        
    def convert(self):
        try:
            frm = self.clicked.get()
            to = self.clicked2.get()
            weight = float(self.screen1.get())
            if frm != to:
                wgt_kg = self.to_kg(frm,weight) 
                if to=="Grams(g)":
                    wgt_kg=round(wgt_kg*1000,ndigits=10)

                elif to=="Tons(t)":
                    wgt_kg=round(wgt_kg*0.001,ndigits=10)

                elif to=="Pounds(lb)":
                    wgt_kg=round(wgt_kg*2.2046226218,ndigits=10)

                elif to=="Ounces(oz)":
                    wgt_kg=round(wgt_kg*35.2739619496,ndigits=10)

                elif to=="Quintal":
                    wgt_kg=round(wgt_kg*0.01,ndigits=10)

                self.screen2["text"] = str(wgt_kg)
            else:    
                self.screen2["text"] = str(weight)
        except ValueError:
            messagebox.showerror("Input Error","Incorrect input!!!")
            
    def to_kg(self,frm,weight):
        if frm=="Grams(g)":
            weight*= 0.001
            return weight    
        elif frm=="Tons(t)":
            weight*= 1000
            return weight
        elif frm=="Pounds(lb)":
            weight*= 0.45359237
            return weight 
        elif frm=="Ounces(oz)":
            weight*= 0.0283495231
            return weight
        elif frm=="Quintal":
            weight*=100
            return weight
        else:
            return weight
```

```python
from tkinter import *
from tkinter import messagebox, simpledialog
from tkinter import ttk as ttk
from datetime import datetime

class Age_calculator:
    def __init__(self,root):
        self.root = root
        self.root.title("Age Calculator")
        self.root.geometry("400x200+400+150")
        self.root.resizable(False,False)
        self.root.config(bg="#cbc1e1")
        self.date = datetime.today().strftime("%d/%m/%Y")
        self.lbl1 = Label(root, text = "Today's Date:",font="Times 13 bold italic",bg="#cbc1e1",anchor="e")
        self.lbl2 = Label(root, text = self.date,font="Times 13 bold italic",fg= "red",bg="#cbc1e1")
        self.lbl3 = Label(root, text = "Enter your Date of Birth:",font="Times 13 bold",bg="#cbc1e1",anchor="w")
        self.lbl4 = Label(root, text = "(in dd/mm/yyyy format)",fg= "red", font = "Ariel 10",bg="#cbc1e1",anchor="ne")
        self.enter = Entry(root,font = "Calibre 15 bold",fg= "green",width = 15,relief="sunken",bd=5)
        self.lbl5 = Label(root,text= "Your Age:",font ="Times 15 bold",height=1,bg="#cbc1e1")
        self.lbl6= Label(root,font ="Calibre 15 bold",fg="green",height=1,width=22,relief="ridge",bd=10)
        self.cal = Button(root,text="Calculate",font="Calibre 9 bold",relief = "raised",width = 9,command=self.age_cal)
        self.clear = Button(root,text="Clear",font="Calibre 9 bold",relief = "raised",width = 9,command=self.clearall)
        self.cancel = Button(root,text="Cancel",font="Calibre 9 bold",relief = "raised",width = 9,command = self.finish)
        self.lbl1.grid(row=0,column=0,sticky=W+E,padx=2)
        self.lbl2.grid(row=0,column=1,sticky=W,padx=2)
        self.lbl3.grid(row=1,column=0,columnspan=2,sticky=W,padx=2)
        self.lbl4.grid(row=2,column=0,sticky=W+N,columnspan=2,padx=2)
        self.enter.grid(row=1,column=2,columnspan=1,sticky=W)
        self.lbl5.grid(row=3,column=0,sticky=W+E,pady=20)
        self.lbl6.grid(row=3,column=1,sticky=W,pady=20,columnspan=3,padx=2)
        self.cal.grid(row=4,column=0,sticky=E,padx=2)
        self.clear.grid(row=4,column=1,sticky=W+E,padx=2)
        self.cancel.grid(row=4,column=2,sticky=W,padx=2)
        
    def age_cal(self):
        try:
            dob= datetime.strptime(self.enter.get(),'%d/%m/%Y')
            years = ((datetime.today()-dob).days/365.2425)
            yearsInt = int(years)
            
            months=(years-yearsInt)*12
            monthsInt=int(months)

            days=(months-monthsInt)*(365.2425/12)
            daysInt=int(days)
            self.lbl6["text"]=f"{yearsInt} years {monthsInt} months {daysInt} days"
        except Exception:
            messagebox.showerror("Invalid Input","Please enter DOB in correct format")
    
    def clearall(self):
        self.enter.delete(0,END)
        self.lbl6["text"]=""
        
    def finish(self):
        ans = messagebox.askyesno("Quitting!!","Do you want to close this program")
        if ans==True:
            self.root.destroy()
            
```

### *Digital Clock*

```python
from tkinter import *
from datetime import datetime
class digital_clock:
    def __init__(self,root):
        self.root = root
        self.root.title("Digital Clock")
        self.root.geometry("520x200+400+150")
        self.root.resizable(False,False)
        self.root.config(bg="#cbc1e1")
        self.lbl = Label(self.root,font="Cambria 60 bold",fg="Red",bg="lightblue",relief="ridge",height=1,width=10,bd=15)
        self.lbl.grid(row=0,column=0,sticky=N+S+E+W,pady=50,padx = 15)
        
        def clock():
            time=datetime.today().strftime("%I:%M:%S %p")
            self.lbl["text"] = time
            self.lbl.after(200,clock)
        
        clock()
```

### *Covid-19 Data Representation*

```python
from tkinter import *
from tkinter import messagebox, simpledialog,filedialog
from tkinter import ttk as ttk
import pandas as pd

def browse():
    file_name= filedialog.askopenfilename(title="Select you data file", filetype=[("csv file","*.csv"),("excel file","*.xlxs")])
    if file_name!='':
        file_label["text"] = file_name
    else:
        messagebox.showerror("Invalid File Selection","No file selected")

def load_data():
    try:
        file_path = r"{}".format(file_label["text"])
        if file_path.endswith(".csv"):
            df=pd.read_csv(file_path)
        elif file_path.endswith(".xlsx"):
            df=pd.read_excel(file_path)
        else:
            raise ValueError("Invalid File type")
        
        data["column"]= list(df.columns)    
        for column in data["column"]:
            data.column(column, anchor="w")
            data.heading(column, text=column)
            
        df_rows = df.to_numpy().tolist()
        for row in df_rows:
            data.insert("","end",text = "1",values=row)
    except ValueError:
        messagebox.showerror("Invalid File","Not valid file type")
    

def clear():
    data.delete(*data.get_children())

def cancel():
    ans = messagebox.askyesno("Quitting","Do you want to quit this program?")
    if ans==True:
        root.destroy()


root = Tk()
root.title("Covid19 Data")
root.geometry("800x500+300+100")
root.resizable(0,0)
top_frame=Frame(root,height=1,relief="ridge",bd=10,bg="yellow")
middle_frame=Frame(root,relief="ridge",bd=10,bg="blue",height=350)
bottom_frame=Frame(root,height=2,relief="ridge",bd=10,bg="lightgreen")
top_frame.pack(fill=X,padx=10,side=TOP)
middle_frame.pack(fill="both",padx=10,expand=True)
bottom_frame.pack(fill=X,padx=10,side=BOTTOM)

#-------------Top Frame--------------
main_label=Label(top_frame, text="Covid-19 Cases Countrywise",bg="yellow",font=("Verdana",20,"bold"))
main_label.pack()

#------------Middle frame------------
data = ttk.Treeview(middle_frame,show="headings",height=17)
data.pack(fill=BOTH,expand=True)
datascrolly = Scrollbar(data,orient = "vertical",command = data.yview)
datascrollx = Scrollbar(data,orient = "horizontal",command = data.xview)
data.configure(xscrollcommand = datascrollx.set,yscrollcommand = datascrolly.set)
datascrollx.pack(side="bottom",fill=X)
datascrolly.pack(side="right",fill=Y)

#------------Bottom Frame------------
file_label = Label(bottom_frame,width = 50,relief="sunken",bd=5,font="Times 10 bold")
but_frame = Frame(bottom_frame,width = 50,relief="flat",bg="lightgreen")
browse = Button(but_frame,text="Browse",command=browse,font="Calibri 10 bold",relief="raised",bd=5,width=10)
load = Button(but_frame,text="Load Data",command=load_data,font="Calibri 10 bold",relief="raised",bd=5,width=10)
clear = Button(but_frame,text="Clearall",command=clear,font="Calibri 10 bold",relief="raised",bd=5,width=10)
cancel = Button(but_frame,text="Cancel",command=cancel,font="Calibri 10 bold",relief="raised",bd=5,width=10)
file_label.grid(row=0,column=0,columnspan=3,sticky=E+W,padx=2)

but_frame.grid(row=0,column=3)
browse.pack(side=LEFT,padx=3)
load.pack(side=LEFT,padx=3)
clear.pack(side=LEFT,padx=3)
cancel.pack(side=LEFT,padx=3)
root.mainloop()
```

### *Notepad*

```python
from tkinter import *
from tkinter import messagebox, simpledialog,filedialog
import os

def newFile():
    global file
    root.title("Untitled - Shivansh's Notepad")
    file=None
    TextArea.delete(1.0,END)

def openFile():
    global file
    file = filedialog.askopenfilename(title="Select your file",
                                      filetypes=[("All files","*.*"),("Text Document","*.txt")])
    if file=="":
        file=None
    else:
        root.title(os.path.basename(file)+"- Shivansh's Notepad")
        TextArea.delete(1.0,END)
        f= open(file,"r")
        TextArea.insert(1.0,f.read())
        f.close()

def saveFile():
    global file
    if file==None:
        file = filedialog.asksaveasfilename(initialfile="Untitled.txt",
                                            defaultextention=".txt",
                                            filetypes=[("All files","*.*"),("Text Document","*.txt")])
        if file =="":
            file=None
        else:
            f=open(file,"w")
            f.write(TextArea.get(1.0,END))
            f.close()                                     
            root.title((os.path.basename(file)+"- Shivansh's Notepad"))
    else:
        f=open(file,"w")
        f.write(TextArea.get(1.0,END))
        f.close()                                             

def exit():
    ans = messagebox.askyesno("Closing!!","Do you want to exit?")
    if ans==True:
        root.destroy()

def copy():
    TextArea.event_generate(("<<Copy>>"))

def cut():
    TextArea.event_generate(("<<Cut>>"))

def paste():
    TextArea.event_generate(("<<Paste>>"))

def about():
    messagebox.showinfo("About Notepad","Notepad Created by Shivansh Kaushal.")

root = Tk()
root.title("Untitled - Shivansh's Notepad")
root.wm_iconbitmap("3.ico")
root.geometry("800x500")
TextArea  = Text(root, font = "Lucida 13")
file = None

MenuBar = Menu(root)
Filebar = Menu(MenuBar,tearoff = False)
Filebar.add_command(label="New",command=newFile)
Filebar.add_command(label="Open",command=openFile)
Filebar.add_command(label="Save",command=saveFile)
Filebar.add_separator()
Filebar.add_command(label="Exit",command=exit)
MenuBar.add_cascade(label="File",menu=Filebar)

Editbar = Menu(MenuBar,tearoff = False)
Editbar.add_command(label="Copy",command=copy)
Editbar.add_command(label="Cut",command=cut)
Editbar.add_command(label="Paste",command=paste)
MenuBar.add_cascade(label="Edit",menu=Editbar)

Helpbar = Menu(MenuBar,tearoff = False)
Helpbar.add_command(label="About Notepad",command=about)
MenuBar.add_cascade(label="Help",menu=Helpbar)

root.config(menu=MenuBar)

xScrolly = Scrollbar(root,orient = "horizontal",command=TextArea.xview)
yScrolly = Scrollbar(root,orient = "vertical",command=TextArea.yview)
TextArea.config(yscrollcommand=yScrolly.set,xscrollcommand=xScrolly.set)
xScrolly.pack(side="bottom",fill=X)
yScrolly.pack(side="right",fill=Y)
TextArea.pack(fill=BOTH,expand=True)

root.mainloop()
```
