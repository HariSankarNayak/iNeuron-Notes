```python
import tkinter
root = tkinter.Tk()
root.geometry("200x600")
lbl1=tkinter.Label(root,text="Welcome To Python",bg="red")
lbl2=tkinter.Label(root,text="Python Rocks!",bg="blue")
lbl2.pack()
lbl1.pack()
root.mainloop()
```

As discussed previously , the 2 Labels we have added will be added to the window in vertical order.

Another point to notice is that there is no space between their edges.

To change both these behaviors we must perform layouting in Python

The term layouting means arranging the UI components as per our requirement.

In simple words it means determining size and position of a widget

To perform layouting , Python provides us 3 methods:

pack()
place()
grid() 


The method pack() is the easiest to use .


Instead of having to declare precisely where a widget should appear on the display screen, we can declare the positions of widgets with the pack( ) method relative to each other. 


The pack() method takes care of the details. 


```python
import tkinter
root = tkinter.Tk()
#root.geometry("200x200")
l1=tkinter.Label(root, text="Red Sun", bg="red", fg="white")
l2=tkinter.Label(root, text="Green Grass", bg="green", fg="black")
l3=tkinter.Label(root, text="Blue Sky", bg="blue", fg="white")
l1.pack()
l2.pack()
l3.pack()
root.mainloop()
```

In the previous example we used pack() without any options. 


So pack( ) had to decide which way to arrange the labels. 


As we can see, it has done 2 things:
Placed the label widgets vertically below each other and centre them. 
Each label has been given the size of the text. 

If we want to make the widgets as wide as the parent widget, we have to use the fill=tkinter.X option


The fill argument has four possible values:


NONE (default)

X (fill only horizontally)

Y (fill only vertically)

BOTH (fill both horizontally and vertically).


```python
import tkinter
root = tkinter.Tk()
l1=tkinter.Label(root, text="Red Sun", bg="red", fg="white")
l2=tkinter.Label(root, text="Green Grass", bg="green", fg="black")
l3=tkinter.Label(root, text="Blue Sky", bg="blue", fg="white")
l1.pack(fill=tkinter.X)
l2.pack(fill=tkinter.X)
l3.pack(fill=tkinter.X)
root.mainloop()
```

The term padding means leaving space in and around the widget.


The pack() method knows four padding options, i.e. internal and external padding and padding in x and y direction.


Padding in x and y direction means padding outside the widget and internal and external padding means padding between the contents and the border of the widget


```python
import tkinter
root = tkinter.Tk()
l1=tkinter.Label(root, text="Red Sun", bg="red", fg="white")
l2=tkinter.Label(root, text="Green Grass", bg="green", fg="black")
l3=tkinter.Label(root, text="Blue Sky", bg="blue", fg="white")
l1.pack(fill=tkinter.Y,padx=10)
l2.pack(fill=tkinter.Y,padx=10)
l3.pack(fill=tkinter.Y,padx=10) 
root.mainloop()
```

As we can see the padding of 10 has occurred on both the  x –axis of the widget.

If however we want it to occur only on one side , we can pass the values as tuple

To pack widgets side by side, we use the side option.

It’s possible values are:

TOP (default)

BOTTOM

LEFT

RIGHT


```python
import tkinter
root = tkinter.Tk()
l1=tkinter.Label(root, text="Red Sun", bg="red", fg="white")
l2=tkinter.Label(root, text="Green Grass", bg="green", fg="white")
l3=tkinter.Label(root, text="Blue Sky", bg="blue", fg="white")
l1.pack(padx=5,pady=5,side=tkinter.LEFT)
l2.pack(padx=5,pady=5,side=tkinter.LEFT)
l3.pack(padx=5,pady=5,side=tkinter.LEFT)
root.mainloop() 
```

The method place( ) allows us to explicitly set the position and size of a widget, either in absolute terms, or relative to another widget. 


When using this layout manager the size and position of the widgets do not change while resizing the window.

Syntax:
widget.place(options)

```python
import tkinter
root = tkinter.Tk()
root.geometry('200x200+100+200')
l1=tkinter.Label(root, text="Name:")
l2=tkinter.Label(root, text="Password:")
e1=tkinter.Entry(root)
e2=tkinter.Entry(root)
b1=tkinter.Button(root,text="Login")
b2=tkinter.Button(root,text="Quit")
l1.grid(row=0,column=0)
e1.grid(row=0,column=1)
l2.grid(row=1,column=0)
e2.grid(row=1,column=1)
b1.grid(row=2,column=0)
b2.grid(row=2,column=1)
root.mainloop()
```

If we closely observe the above output , we will notice that the first Label with the text “Name:” is center align.


This is because the method grid( ) places the component in it’s cell with center aligned.


To change this behavior we must use the sticky option which tells Python to set the component aligned towards N,E,W or S edges of the cell.


```python
import tkinter
root = tkinter.Tk()
root.geometry('200x200+100+200')
l1=tkinter.Label(root, text="Name:")
l2=tkinter.Label(root, text="Password:")
e1=tkinter.Entry(root)
e2=tkinter.Entry(root)
b1=tkinter.Button(root,text="Login")
b2=tkinter.Button(root,text="Quit")
l1.grid(row=0,column=0,sticky=tkinter.W)
e1.grid(row=0,column=1)
l2.grid(row=1,column=0)
e2.grid(row=1,column=1)
b1.grid(row=2,column=0)
b2.grid(row=2,column=1)
root.mainloop()
```

Write an app to generate the following UI:
(This is called CheckButton in tkinter and to create them we write:
c1=tkinter.CheckButton(root,text=“Keep me logged in”))


Before we design the required layout , can you tell how many rows and columns the window has ?

Answer is 4 rows and 3 columns

But in the first and second row we only have 2 components ,so we will have to tell Python to stretch the Entry components so that they cover up 2 columns horizontally.

This is done using the attribute columnspan

Same thing needs to be done with the CheckButton 


Another important point to understand is that the Buttons are taking up entire cell width in which they are placed.

This is done by assigning the value tkinter.W+tkinter.E to the sticky attribute


```python
import tkinter
root = tkinter.Tk()
root.geometry('200x200+100+200')
l1=tkinter.Label(root, text="Name:")
l2=tkinter.Label(root, text="Password:")
e1=tkinter.Entry(root)
e2=tkinter.Entry(root)
b1=tkinter.Button(root,text="Login")
b2=tkinter.Button(root,text="Quit")
c1=tkinter.Checkbutton(root,text="Keep me logged in")
l1.grid(row=0,column=0,sticky=tkinter.E)
e1.grid(row=0,column=1,columnspan=2)
l2.grid(row=1,column=0)
e2.grid(row=1,column=1,columnspan=2)
c1.grid(row=2,column=0,columnspan=2)
b1.grid(row=3,column=1,sticky=tkinter.W+tkinter.E)
b2.grid(row=3,column=2,sticky=tkinter.W+tkinter.E)
root.mainloop()
```
