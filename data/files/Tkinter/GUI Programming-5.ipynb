If we want to ask the user for a single data value, either a string, integer, or floating point value, we can use the tkinter.simpledialog module. 


The user can enter the requested value and hit “OK”, which will return the entered value. If the user hits “Cancel,” then None is returned.


To produce an input dialog we can call either of the following three functions present in the tkinter.simpledialog module.

These are :

askstring()

askinteger()

askfloat()


The function askstring( ) accepts 2 positional arguments and 2 optional arguments:
Syntax:
simpledialog.askstring(title, prompt [,options])

title: The title of the dialog box

prompt: The message to be displayed

initialvalue: Initial value, if any. Default is an empty string.

parent: Which window to place the dialog on top of.

The askinteger() and askfloat() functions is similar to askstring( ), but they only accept integers and float values, respectively. 

All the parameters are same as askstring( ) with 2 new options called  minvalue and maxvalue  to limit the input range: 

Syntax:
simpledialog.askinteger(title, prompt [,options])
simpledialog.askfloatr(title, prompt [,options])


Following are important points about the behaviour of these functions:

If the input value is a valid int or float value and within the provided minvalue and maxvalue range then the value is returned and dialog box is removed

If the input value is either invalid or  out of the given range an error dialog is displayed and the user is forced to input again

If the user presses cancel button then None is returned and the dialog box is removed


```python
from tkinter import *
from tkinter import messagebox
from tkinter import simpledialog
def accept():
    num=simpledialog.askinteger("Input","Enter voter's age",minvalue=18,maxvalue=122)
    messagebox.showinfo("Hello","You entered "+str(num))
root = Tk()
root.geometry("200x200")
btn=Button(root,text="Click Me",command=accept)
btn.pack()
root.mainloop()
```

```python
from tkinter import *
from tkinter import colorchooser, messagebox

def showopen():
    chosen_color=colorchooser.askcolor(color="#ffff00",title="Select a color")
    if chosen_color[0] is None:
        messagebox.showinfo("Color","No color selected")
    else:
        root.configure(bg=chosen_color[1])
root = Tk()
root.geometry("200x200")
btn=Button(root,text="Choose Color",command=showopen)
btn.pack()
root.mainloop()
```
