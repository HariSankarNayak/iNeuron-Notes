```python
import pandas as pd
df = pd.read_csv("bank/bank.csv", sep=';')
```

```python
df
```

```python
# Q1 : how many campaign available in this dataset .
total_campaign = sum(df['campaign'])
print("Total Number of campaign = ", total_campaign)
```

```python
# Q2 : how many users do we have with housing and personal loan . 
ppl_with_hous_pers_loan = len(df[(df['housing'] == 'yes') & (df['loan'] == 'yes')])
print("Number of people with Housing & PErsonal loan = ", ppl_with_hous_pers_loan)
```

```python
# Q3 : how many person do we have whose age is 60+
ppl_with_age_gt_60 = len(df[df['age'] >= 60])
print("Number of people with age greater than 60 = ", ppl_with_age_gt_60)
```

```python
# Q4 : in which month we have trageted most of the customer . 
list_months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
list_count = []
for i in list_months:
    count = len(df[df['month'] == i])
    list_count.append(count)
print(list_count)

max_val_index = list_count.index(max(list_count))
print("Month in which most customers targeted = ", list_months[max_val_index])
```

```python
# Q5 : which mode of call is giving you more result
Cellular_Success = len(df[ (df['contact'] == 'cellular') & ( df['poutcome'] == 'success') ])
Telephone_Success = len(df[ (df['contact'] == 'telephone') & ( df['poutcome'] == 'success') ])
print(Cellular_Success, Telephone_Success)
if(Cellular_Success > Telephone_Success):
    print("Cellular mode of call is more successfull")
else:
    print("Telephone mode of call is more successfull")
```

```python
#Q6 : how many enterpeures do we have in this list 
entrepreneur_count = len(df[df['job'] == 'entrepreneur'])
print("Number of entrepreneur = ", entrepreneur_count)
```

```python
#Q7 : how many customers do we have with negative balance 
Cust_Neg_Bal = len(df[df['balance'] < 0])
print("Number of customers with negative balance = ", Cust_Neg_Bal)
```

```python
#Q8 : prepare a group of data based on education level . 
primary_df = df[df['education'] == 'primary']
print(primary_df)

secondary_df = df[df['education'] == 'secondary']
print(secondary_df)

tertiary_df = df[df['education'] == 'tertiary']
print(tertiary_df)
```
