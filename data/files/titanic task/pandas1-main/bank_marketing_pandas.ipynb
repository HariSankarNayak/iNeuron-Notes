```python
import pandas as pd
```

```python
df = pd.read_csv('pandas_assessment1/bank.csv', delimiter=';')
```

```python
df
```

```python
# 1. how many campaigns are available in this dataset
number_of_campaigns = df['campaign'].unique().size
print('Number of campaigns available:', number_of_campaigns)
```

```python
# 2. how many users do we have with housing and personal loan
users_with_housing_and_loan = len(df[(df.housing == 'yes') & (df.loan=='yes')])
print('Number of users with housing and loans:', users_with_housing_and_loan)
```

```python
# 3. how many persons do we have with age 60+
users_aged_60plus = len(df[df.age>60])
print('Number of users aged 60+:', users_aged_60plus)
```

```python
#4. in which month we have targeted most of the customers
themonth = df.month.value_counts().index[0]
occurences = df.month.value_counts().may
print(f'Month we targeted most of the customers: {themonth.title()} \
with {occurences} customers targeted')
```

```python
def number_of_calls(index):
    return df.contact.value_counts().values[index]

def calls_outcome(outcome, index):
    grouping = df[(df.contact != None) & (df.poutcome == outcome)].groupby(df.contact).size()
    return grouping.values[index]
```

```python
# 5. which mode of call is giving you more results

cellular_success_rate = calls_outcome('success', 0) / number_of_calls(0) * 100
telephone_success_rate = calls_outcome('success', 1) / number_of_calls(1) * 100

print('Cellular success rate:', round(cellular_success_rate, 2))
print('Telephone success rate:', round(telephone_success_rate, 2))
# The problem with the data set is that we have no idea about the outcomes of most of the
# calls. 
number_of_calls_with_unknown_outcomes = calls_outcome('unknown', 0)
print(f'It\'s however important to note that we actually have {number_of_calls_with_unknown_outcomes} calls we know nothing about the outcomes')
print()
print(f'At the same time {number_of_calls(0)} calls were made using cellular.')
print(f'From that number, {calls_outcome("success", 0)} were successfull')
```

```python
# 6. how many entrepreneurs do we have in the list
number_of_entrepreneurs = df.job[(df.job == 'entrepreneur') | (df.job == 'self-employed')].size
number_of_strict_entrepreneurs = df.job[(df.job == 'entrepreneur')].count()
print('Number of users with the label "Entrepreneur":', number_of_strict_entrepreneurs)
print(f'If we count self-employed workers as entrepreneurs, there are {number_of_entrepreneurs} entrepreneurs.')
```

```python
# 7. prepare a group of data based on education level
edugroups = df.groupby(df.education)

primary_level_group = edugroups.get_group('primary')
secondary_level_group = edugroups.get_group('secondary')
tertiary_level_group = edugroups.get_group('tertiary')
unknown_level_group = edugroups.get_group('unknown')
edugroups.describe()
```
