```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()
```

```python
# 4. What is the name of a person who is the eldest one?
df[df['Age'] == max(df['Age'])]['Name']
```

```python
# 5. How many passenger are there in the first, second and third class?
df['Pclass'].value_counts()
```

```python
# 6. How many people whose name start with "S"?
len(df[df['Name'].str.startswith('S')]["Name"])
```

```python
# 7. Try to create a new column which is a summation of "SibSp" and "parch"?
df['SibSp-Parch'] = df["SibSp"] + df["Parch"]
df
```

```python
# 8. How many people have the age bellow 25?
len(df[df["Age"] < 25])
```

```python
# 9. How many people died whose age less than 40?
len(df.loc[(df['Survived'] == 1) & (df["Age"] < 40)])
```

```python
 # 10. From a Cabine column, seperate text and numeric value
for val in df["Cabin"]:
    if not pd.isna(val):
        char = ''
        num = ''
        for ch in val:
            if ch.isdigit():
                num += ch
            else:
                char += ch
        print(f"For `{val}`: char `{char}` and number: `{num}`")
```
