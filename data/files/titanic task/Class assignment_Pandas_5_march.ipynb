```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
df.head()
```

```python
#Q1 how many male and female
df['Sex'].value_counts()
```

```python
#Q2 how many survived
survived_count = df[df['Survived'] == 1]['Survived'].count()
print('no of survived:',survived_count)
```

```python
#Q3 how many casuality
causality_count = df[df['Survived'] == 0]['Survived'].count()
print('no of casualities:',causality_count)
```

```python
 #Q4 what is name of a person who is the eldest
eldest_person = df[df['Age'] == max(df['Age'])]['Name']
print('eldest person:',eldest_person)
```

```python
#Q5. how many passenger do we have in first , second and third class 
df['Pclass'].value_counts()
```

```python
#Q6. how many person we have whose name starts with "s"
name_strts_S_count = df[df['Name'].str.startswith('S')]['Name'].count()
print("no of persons whose name starts with 's':", name_strts_S_count)
```

```python
#Q7. try to create a new column which is a summation  of "SibSp" and "parch"
df['SibSp_Parch'] = df['SibSp'] + df['Parch']
df.head()
```

```python
#Q8 how many person do we have below age of 25
below_25_count = df[df['Age'] < 25]['Age'].count()
print('no of people below 25yr age:',below_25_count)
```

```python
#Q9 how many person died whose age was less than 40
df1 = df[df['Age'] < 40]
died_below_40_count =df1[df1['Survived'] == 0]['Survived'].count()
print('no of people died below 40yr age:',died_below_40_count )
```

```python
#Q10 from a cabin column separate text and numeric value
df['Cabin_number'] = df['Cabin'].str.replace('([A-Za-z]+)', '')
df['Cabin_text'] = df['Cabin'].str.extract('([A-Za-z]+)')
df
```
