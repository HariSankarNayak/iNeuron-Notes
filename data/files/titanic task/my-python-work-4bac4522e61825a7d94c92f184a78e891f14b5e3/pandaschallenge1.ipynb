```python
import pandas as  pd 
df=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head()
```

```python
df.columns
```

```python

#1 . Find out how many male and female passenger was onboarded .

print("no. of males",df[df['Sex']=='male'].shape[0])
print("no. of females",df[df['Sex']=='female'].shape[0])
```

```python
#how many survived we have .
#how many casuality we have 

print("no. of survived",df[df['Survived']==1].shape[0])
print("no. of casualities",df[df['Survived']==0].shape[0])
```

```python
#4 . what is name of a person who is the eldest one .

print("eldest person",df.loc[df['Age']==df['Age'].max(),'Name'].values)
```

```python
#5 . how many passenger do we have in first , second and third class 

print("distribution of passengers in different class\n",df['Pclass'].value_counts())
```

```python
#6. how many person we have whose name starts with "s"

count=0
for i in df['Name'].values:
    if i.startswith("S")==True:
        count+=1
print("person we have whose name starts with 's'",count)        
```

```python
#7. try to create a new column which is a summation  of "SibSp" and "parch"

summation=df['SibSp']+df['Parch']
summation
df['Sib+Parch']=summation
df
```

```python
#how many person do we have below age of 25
df[df['Age']<25].shape[0]
```

```python
#how many person died whose age was less then 40 
df[(df['Age']<40)&(df['Survived']==0)].shape[0]
```
