```python
user_input=int(input("enter the number to be checked"))
if user_input%2==0:
    print("number is even")
else:
    print("number is odd")
    
```

```python
user_input=int(input("enter the year to be checked"))
if user_input%100==0:
    if user_input%400==0:
        print("the year is a leap yr")
    else:
        print("the year is not a leap yr")
else:
    if user_input%4==0:
        print("the year is a leap yr")
    else:
        print("the year is not a leap yr")
```

```python
user_input=int(input("enter the year to be checked"))
rem=[user_input%i for i in range(2,user_input)]
if 0 not in rem:
    print("the number is prime")
else:
    print("the number is composite")
```
