# List of all values in a dictionary

```python
d = {'a': 1, 'b': 2, 'c': {'c': 3, 'd': 4 , 'e' : {'m': 8, 'n': 9}}, 'e': 5, 'f':{'f':6,'g':7}}

def dict_val(a,lst = []):
    '''This program lists all values inside a dictionary input'''
    for i in a.values():
        if type(i) == dict:
            dict_val(i)
        else:
            lst.append(i)
    return(lst)

dict_val(d)
```

# Concatenating lists of user's choice

```python
def list_concat():
    '''This function concatenates items of multiple lists received from user'''
    n = int(input("How many lists do you want concatenated?? "))
    lst = []
    for i in range(n):
        k = int(input('How many elements do you want in list No.%d ' %(i+1)))
        for j in range(k):
            l = input('Input each element and press Enter:')
            lst.append(l)

    return(print('Your concatenated list is ',lst))

list_concat()
```

```python
import cv2

def show_vid(video_path):
    video = cv2.VideoCapture(video_path)
    
    while(video.isOpened()):
        # Capturing the video fram by frame
        ret, frame = video.read()
        if ret == True:
            cv2.imshow('Frame',frame)# Displaying the current frame
            if cv2.waitKey(25) & 0xFF == ord('x'): # Press 'x' to exit the video
                break
        else:
            break
    video.release()
    cv2.destroyAllWindows()
```

# List all primitive elements from a given input

```python
def concatall(*args):
    '''This function accepts arguments from the user and then displays each primitive element'''
    l = []
    intlst = []
    fltlst = []
    strlst = []
    boolst = []
    cmplxlst = []
    d = {}
    for i in args:
        if type(i) == list:
            l+=i
        elif type(i) == int:
            intlst.append(i)
        elif type(i) == float:
            fltlst.append(i)
        elif type(i) == str:
            strlst.append(i)
        elif type(i) == bool:
            boolst.append(i)
        elif type(i) == complex:
            cmplxlst.append(i)
        elif type(i) == dict:
            d.update(i)
    return (print('List of lists',l),
            print('List of integers',intlst),
            print('List of floats',fltlst),
            print('List of strings',strlst),
            print('List of Booleans',boolst),
            print('List of complex nos.',cmplxlst),
            print('List of dicts',d)
            )

concatall([11,22],3,4,'hello',{'a':12,'b':34},'world',[89,67,45],4+6j,{'c':56,'d':112,'e':87})
```

# Show ip address

```python
def showip():
    '''This function displays the IP address of the system'''
    
    import socket
    name = socket.gethostname()
    ip_add = socket.gethostbyname(name)
    print("System name is ",name,'and ip address is ',ip_add)

showip()
```
