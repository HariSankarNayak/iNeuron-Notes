```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# to visualize all the columns in the dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
dataset.head()
```

```python
X = dataset.drop(["Id","SalePrice"], axis=1)
y = dataset["SalePrice"]
print(X)
print(y.shape)
```

```python
X=dataset.iloc[:,1:81] # all the columns except the first and the last one . As the first column is Id (not a feature) and the last column is SalePrice (target variable)
y=dataset.iloc[:,-1] # the last column
```

```python
## The train_test_split() function takes several parameters, including the features (X), the target variable (y), the test size (test_size), and a random state (random_state) for reproducibility.
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1)
 ## train data is used to train the model and test data is used to evaluate the model on unseen data. 
 # The test size is set to 0.1, which means that 10% of the data will be used for testing and 90% will be used for training.
```

```python
#Missing Values
features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>=1]
for feature in features_nan:
    print("The feature is {} and missing values in {}%".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
## Replacing the missing values in categorical features with 'Missing'
## how to get categorical features in the dataset ? Ans: we can use the dtypes of the columns to get the categorical features. The dtypes of the columns will be 'O' for categorical features.

categorical_features=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
```

```python
categorical_features
```

```python
## Replacing the missing values in categorical features with 'Missing'
data=dataset.copy()
def replace_cat_features(dataset,categorical_features):
    data[categorical_features]=data[categorical_features].fillna('Missing')
    return data
```

```python
replace_cat_features(data,categorical_features)
```

```python
dataset['LotFrontage'].mode()
```

```python
## Replacing the missing values in categorical features with mode
data=dataset.copy()
def replace_cat_features(dataset,categorical_features):
    data[categorical_features]=data[categorical_features].fillna(data[categorical_features].mode())
    return data
```

```python
replace_cat_features(data,categorical_features)
```

```python
numerical_features_nan=[feature for feature in dataset.columns if dataset[feature].dtypes!='O' and
                       dataset[feature].isnull().sum()>=1]
```

```python
numerical_features_nan
```

```python
for feature in numerical_features_nan:
    ## since there are outliers we are goign to replace wwith median
    median_values=dataset[feature].median()
    ## create a new feature to capture nan value
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_values,inplace=True)
```

```python
##temporal variables(datet time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature+' age of house']=dataset['YrSold']-dataset[feature]
    
```
